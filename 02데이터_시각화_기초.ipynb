## 기본 그래프 그리기

```python
import matplotlib.pyplot as plt
plt.plot([10, 20, 30, 40])
plt.show()
```

plot()함수에 입력된 리스트는 y축 값으로 입력되었으며 x축 값은 자동으로 0부터 1씩 증가하는 정수로 입력되었다.

```python
plt.plot([1, 2, 3, 4], [12, 43, 25, 15])
plt.show()
```

## 그래프에 옵션 추가하기

```python
# 그래프 제목넣기
plt.title('plotting')
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
# 그래프에 범례 넣기
plt.title('legend')
plt.plot([10, 20, 30, 40], label = 'asc')
plt.plot([40, 30, 20, 10], label = 'desc')
plt.legend()
plt.show()
```

```python
# 그래프 색상 바꾸기
plt.title('color')
plt.plot([10, 20, 30, 40], color = 'skyblue', label = 'skyblue')
plt.plot([40, 30, 20, 10], 'pink', label = 'pink')
plt.legend()
plt.show()
```

```python
# 그래프 선 모양 바꾸기
plt.title('linestyle')
plt.plot([10, 20, 30, 40], color = 'r', linestyle = '--', label = 'dashed')
plt.plot([40, 30, 20, 10], color = 'g', ls = ':', label = 'dotted')
plt.legend()
plt.show()
```

```python
# 마커 모양 바꾸기
plt.title('marker')
plt.plot([10, 20, 30, 40], 'r.', label = 'circle')
plt.plot([40, 30, 20, 10], 'g^', label = 'triangle up')
plt.legend()
plt.show()
```

## 내 생일의 기온 변화를 그래프로 그리기

```python
# 데이터 읽어오기
import csv 
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', encoding = 'cp949')
data = csv.reader(f)

next(data)  # 맨 윗줄 헤더 부분은 제외시키기

# 최고 기온 리스트 만들기
result = []

for row in data:
  if row[-1] != '':
    result.append(row[-1])

print(result)
```

```python
# 1983 이후 데이터만, 최고 + 최저 기온 데이터를 그래프로 나타내기
high = []
low = []
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', encoding = 'cp949')
data = csv.reader(f)

next(data)

for row in data:
  if row[-1] != '':
    if int(row[0].split('-')[0]) >= 1983: 
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '15':
        high.append(float(row[-1]))
        low.append(float(row[-2]))

f.close()
plt.plot(high, 'hotpink', label = 'high')
plt.plot(low, 'skyblue', label = 'low')
plt.legend()
plt.show()
```

```python
# 1983 이후 데이터만, 최고 + 최저 기온 데이터를 그래프로 나타내기
high = []
low = []
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', encoding = 'cp949')
data = csv.reader(f)

next(data)

for row in data:
  if row[-1] != '':
    if int(row[0].split('-')[0]) >= 1983: 
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '15':
        high.append(float(row[-1]))
        low.append(float(row[-2]))

f.close()

plt.rc('font', family = 'Malgun Gothic')
plt.title('내 생일의 기온 그래프')

plt.plot(high, 'hotpink', label = 'high')
plt.plot(low, 'skyblue', label = 'low')
plt.legend()
plt.show()
```

### 히스토그램

```python
plt.hist([1, 1, 2, 3, 4, 5, 6, 6, 7, 8, 10])
plt.show()
```

```python
# 주사위 시뮬레이션
import random

dice = []
for i in range(5):
  dice.append(random.randint(1, 6))

plt.hist(dice)
plt.show()
```

```python
# 주사위 100번 던지기
dice = []

for i in range(100):
  dice.append(random.randint(1, 6))

plt.hist(dice, bins = 6)
plt.show()
```

```python
# 기온 데이터를 히스토그램으로 표현하기
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', encoding = 'cp949')
data = csv.reader(f)

next(data)

result = []

for row in data:
  if row[-1] != '':
    result.append(float(row[-1]))

plt.hist(result, bins = 100, color = 'r')
plt.show()
```

```python
from ssl import ALERT_DESCRIPTION_UNEXPECTED_MESSAGE
# 기온 데이터를 히스토그램으로 표현하기
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', encoding = 'cp949')
data = csv.reader(f)

next(data)

aug = []

for row in data:
  month = row[0].split('-')[1]

  if row[-1] != '':
    if month == '08':
      aug.append(float(row[-1]))

plt.hist(aug, bins = 100, color = 'r')
plt.show()
```

```python
# 1월과 8월의 데이터 시각화하기
from ssl import ALERT_DESCRIPTION_UNEXPECTED_MESSAGE
# 기온 데이터를 히스토그램으로 표현하기
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', encoding = 'cp949')
data = csv.reader(f)

next(data)

aug = []
jan = []

for row in data:
  month = row[0].split('-')[1]

  if row[-1] != '':
    if month == '08':
      aug.append(float(row[-1]))
    elif month == '01':
      jan.append(float(row[-1]))

plt.hist(aug, bins = 100, color = 'r')
plt.hist(jan, bins = 100, color = 'b')
plt.show()
```

### 상자 그림(boxplot)  
자료에서 얻어낸 최댓값, 최솟값, 상위 1/4, 2/4(중앙), 3/4에 위치한 값을 보여주는 그래프.

```python
# 기온 데이터를 상자 그림으로 표현하기
result = []

for i in range(13):  # 임의의 데이터 만들기
  result.append(random.randint(1, 1000))
print(sorted(result))

plt.boxplot(result)
plt.show()
```

```python
# 서울 최고 기온 데이터를 상자 그림으로 그리기
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', encoding = 'cp949')
data = csv.reader(f)

next(data)

result = []
for row in data:
  if row[-1] != '':
    result.append(float(row[-1]))

f.close()

plt.boxplot(result)
plt.show()
```

```python
# 1월과 8월의 데이터 시각화하기
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', encoding = 'cp949')
data = csv.reader(f)

next(data)

aug = []
jan = []

for row in data:
  month = row[0].split('-')[1]

  if row[-1] != '':
    if month == '08':
      aug.append(float(row[-1]))
    elif month == '01':
      jan.append(float(row[-1]))

plt.boxplot(aug)
plt.boxplot(jan)
plt.show()
```

히스토그램에서는 추측을 통해 파악할 수 있었던 최솟값, 최댓값, 1/4, .. 값의 위치를 boxplot에서는 비교적 명확하게 파악할 수 있다.  
상자 그림의 위 아래 그려진 동그라미는 이상치 값을 표현한 것이다.  

```python
plt.boxplot([aug, jan])
```

```python
# 최고 기온 데이터를 월별로 구분해서 그래프로 나타내기
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', encoding = 'cp949')
data = csv.reader(f)

next(data)

month = [[] for _ in range(12)]

for row in data:
  if row[-1] != '':
    month_index = int(row[0].split('-')[1]) - 1
    month[month_index].append(float(row[-1]))
    

f.close()
plt.boxplot(month)
plt.show()
```

```python
# 8월 일별 기온 데이터를 상자 그림으로 표현하기
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', encoding = 'cp949')
data = csv.reader(f)

next(data)

aug = [[] for _ in range(31)]

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      day = int(row[0].split('-')[2]) - 1
      aug[day].append(float(row[0-1]))
    
f.close()

plt.style.use('ggplot')  # 그래프 스타일 지정
plt.figure(figsize = (10, 5), dpi = 300)
plt.boxplot(aug, showfliers = False)  # 이상치 값이 보이지 않게 설정
plt.show()
```
