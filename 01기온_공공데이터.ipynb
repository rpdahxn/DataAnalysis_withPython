## -1- 서울이 가장 더웠던 날 찾기

```python
# 문자열로 저장된 최고기온 데이터를 실수 데이터로 변환한다.
for row in data :
  row[-1] = float(row[-1])
  print(row)
```

빈 문자열을 실수 데이터로 변환하는 과정에서 에러가 발생했다.  
누락된 곳에는 대체값을 넣어준다.

```python
for row in data:
  if row[-1] == '':
    row[-1] = -999
  row[-1] = float(row[-1])
  print(row)
```

```python
# 최고 기온 값을 저장할 변수
max_temp = -999
# 최고 기온이 가장 높았던 날짜를 저장할 변수
max_date = ''
```

전체 코드

```python
import csv
f = open('/content/drive/MyDrive/Colab Notebooks/DataAnalysis_withPython/seoul.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

header = next(data)

for row in data:
  if row[-1] == '':
    row[-1] = -999
  row[-1] = float(row[-1])

  if max_temp < row[-1]:
    max_temp = row[-1]
    max_date = row[0]
    print(row[0])

f.close()
print('서울의 최고 기온이 가장 높았던 날은', max_date + '로, ', max_temp,'도 였습니다.')
```
